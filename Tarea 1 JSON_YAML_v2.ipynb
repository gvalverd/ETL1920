### JSON

JSON (JavaScript Object Notation) es un formato ligero de intercambio de datos. Es fácil para los humanos leer y escribir. Es fácil para las máquinas analizar y generar.
JSON se basa en dos estructuras:

Una colección de pares de nombre / valor. En varios lenguajes, esto se realiza como un objeto , registro, estructura, diccionario, tabla hash, lista con clave o matriz asociativa.
Una lista ordenada de valores. En la mayoría de los lenguajes, esto se realiza como una matriz , vector, lista o secuencia.

Estas son estructuras de datos universales. Prácticamente todos los lenguajes de programación modernos los admiten de una forma u otra. Tiene sentido que un formato de datos que sea intercambiable con lenguajes de programación también se base en estas estructuras.

# YAML

YAML significa " YAML no es lenguaje de marcado ", lo que lo denota como flexible y orientado a datos. De hecho, se puede utilizar con casi cualquier aplicación que necesite almacenar o transmitir datos. Su flexibilidad se debe en parte al hecho de que YAML se compone de fragmentos de otros lenguajes


#### Características de YAML
Los objetivos de diseño para YAML son, en prioridad decreciente:

YAML es fácilmente legible por humanos.

Los datos YAML son portátiles entre lenguajes de programación.

YAML coincide con el estructuras de datos nativas de lenguajes ágiles.

YAML tiene un modelo consistente para admitir herramientas genéricas.

YAML admite el procesamiento de una pasada.

YAML es expresivo y extensible.

YAML es fácil de implementar y usar.




## CREACION YAML

```python
data = """
- Datos personales:
    Nombre: Sergio
    Apellido_1: Cañón
    Apellido_2: Laiz
    Edad: 23
    
- Proyecto:
    Asignatura: 
        -ETL:
            Nombre: Práctica 1
            Fecha: Hoy
        -ML:
            Nombre: trabajo X
            Fecha: Ayer
"""

```

```python
print(yaml.load(data))
```

```python
with open('data.yaml', 'w') as outfile:
    yaml.dump(data, outfile, default_flow_style=False)
```

### YAML frente a JSON

Tanto JSON como YAML tienen como objetivo ser formatos de intercambio de datos legibles por humanos. Sin embargo, JSON y YAML tienen diferentes prioridades. El principal objetivo de diseño de JSON es la simplicidad y la universalidad. Por lo tanto, JSON es trivial de generar y analizar, a costa de una menor legibilidad humana. También utiliza un modelo de información de mínimo común denominador, lo que garantiza que cualquier entorno de programación moderno pueda procesar fácilmente cualquier dato JSON.

En contraste, los principales objetivos de diseño de YAML son la legibilidad humana y el soporte para serializando arbitrarioestructuras de datos nativas . Por lo tanto, YAML permite archivos extremadamente legibles, pero es más complejo de generar y analizar. Además, YAML se aventura más allá de los tipos de datos de mínimo común denominador, lo que requiere un procesamiento más complejo al cruzar entre diferentes entornos de programación.

Conclusiones clave

YAML es un lenguaje orientado a datos que tiene características derivadas de Perl, C, HTML y otros lenguajes.

YAML es un superconjunto de JSON que viene con múltiples ventajas integradas, como la inclusión de comentarios, la autorreferencia y la compatibilidad con tipos de datos complejos.
